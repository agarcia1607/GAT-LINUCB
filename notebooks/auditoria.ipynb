```python
from yfinance import download

START_DATE = "2020-01-01"
INTERVAL = "1wk"

df = download(
    tickers=tickers,
    start=START_DATE,
    auto_adjust=False,
    group_by="column",
    progress=True,
    threads=True,
    interval=INTERVAL
)

df
```
